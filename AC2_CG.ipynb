# Transformações Geométricas com Matplotlib
Exercícios AC2 24/09/2025 - Computação Gráfica

```python

import numpy as np
import matplotlib.pyplot as plt

def plot_shapes(original_shape, transformed_shape, title, point=False):
    plt.figure()
    if point:
        plt.scatter(original_shape[0][0], original_shape[0][1], label="Original", c="blue")
        plt.scatter(transformed_shape[0][0], transformed_shape[0][1], label="Transformado", c="red", marker="x")
    else:
        plt.plot(*zip(*original_shape, original_shape[0]), label="Original")
        plt.plot(*zip(*transformed_shape, transformed_shape[0]), label="Transformado", linestyle="--")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.axis("equal")
    plt.show()
```

## Exercício 1: Translação Simples

```python

P = np.array([[2, 3]])
translation = np.array([4, -2])
P_prime = P + translation

print("Novo ponto:", P_prime)
plot_shapes(P, P_prime, "Exercício 1 - Translação", point=True)
```

## Exercício 2: Escala Uniforme

```python

triangle = np.array([[1, 1], [3, 1], [2, 4]])
scale_factor = 2
triangle_scaled = triangle * scale_factor

print("Novos vértices:", triangle_scaled)
plot_shapes(triangle, triangle_scaled, "Exercício 2 - Escala Uniforme")
```

## Exercício 3: Escala Não Uniforme

```python

scale_matrix = np.array([[2, 0], [0, 0.5]])
triangle_nonuniform = triangle @ scale_matrix.T

print("Novos vértices:", triangle_nonuniform)
plot_shapes(triangle, triangle_nonuniform, "Exercício 3 - Escala Não Uniforme")
```

## Exercício 4: Rotação em Torno da Origem

```python

P = np.array([[1, 0]])
theta = np.radians(90)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
P_rotated = P @ rotation_matrix.T

print("Novo ponto:", P_rotated)
plot_shapes(P, P_rotated, "Exercício 4 - Rotação 90°", point=True)
```

## Exercício 5: Rotação de um Quadrado

```python

square = np.array([[1, 1], [1, 4], [4, 4], [4, 1]])
theta = np.radians(-45)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
square_rotated = square @ rotation_matrix.T

print("Novos vértices:", square_rotated)
plot_shapes(square, square_rotated, "Exercício 5 - Rotação 45° Horário")
```

## Exercício 6: Reflexão Simples

```python

P = np.array([[2, 5]])
reflection_matrix = np.array([[-1, 0], [0, 1]])
P_reflected = P @ reflection_matrix.T

print("Novo ponto:", P_reflected)
plot_shapes(P, P_reflected, "Exercício 6 - Reflexão eixo Y", point=True)
```

## Exercício 7: Reflexão de um Triângulo

```python

triangle = np.array([[2, 3], [4, 3], [3, 5]])
reflection_matrix = np.array([[1, 0], [0, -1]])
triangle_reflected = triangle @ reflection_matrix.T

print("Novos vértices:", triangle_reflected)
plot_shapes(triangle, triangle_reflected, "Exercício 7 - Reflexão eixo X")
```

## Exercício 8: Cisalhamento Horizontal

```python

P = np.array([[2, 3]])
k = 2
shear_matrix = np.array([[1, k], [0, 1]])
P_sheared = P @ shear_matrix.T

print("Novo ponto:", P_sheared)
plot_shapes(P, P_sheared, "Exercício 8 - Cisalhamento Horizontal", point=True)
```

## Exercício 9: Composição de Transformações

```python

P = np.array([[3, 2]])

# 1. Translação
P1 = P + np.array([1, -1])

# 2. Rotação 90° anti-horário
theta = np.radians(90)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
P2 = P1 @ rotation_matrix.T

# 3. Escala uniforme fator 2
P3 = P2 * 2

print("Novo ponto:", P3)
plot_shapes(P, P3, "Exercício 9 - Composição de Transformações", point=True)
```

## Exercício 10: Combinação de Transformações em um Retângulo

```python

rectangle = np.array([[1, 1], [5, 1], [5, 3], [1, 3]])

# 1. Translação (-2, 3)
rect1 = rectangle + np.array([-2, 3])

# 2. Escala não uniforme (1.5, 0.5)
scale_matrix = np.array([[1.5, 0], [0, 0.5]])
rect2 = rect1 @ scale_matrix.T

# 3. Reflexão eixo y
reflection_matrix = np.array([[-1, 0], [0, 1]])
rect3 = rect2 @ reflection_matrix.T

print("Novos vértices:", rect3)
plot_shapes(rectangle, rect3, "Exercício 10 - Transformações em Retângulo")
```
